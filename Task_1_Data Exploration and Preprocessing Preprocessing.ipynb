```python
# Import all libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
df = pd.read_csv('Dataset .csv')
df.head()
```

```python
# 1. Explore the dataset
df.shape # Number of rows and columns

```

```python
df.head()  # First few rows
```

```python
df.info()  # Data types and missing values
```

```python
# 2. Handle missing values
# Check for missing values
print(df.isnull().sum())

```

```python
# Handle missing values (example: fill with mean)
# Check for missing values
print(df.isnull().sum())

# Handle missing values based on the specific column and its nature:

# 1. Categorical Columns:
#   - If missing values are significant, consider dropping those rows.
#   - If not, fill with the most frequent category or a new category "Missing".
df['Cuisines'].fillna('Missing', inplace=True)

# 2. Numerical Columns:
#   - Fill with mean, median, or mode depending on the distribution.
#   - For continuous variables, imputation techniques like KNN imputation can be used.
df['Latitude'].fillna(df['Latitude'].mean(), inplace=True)

# 3. Boolean Columns:
#   - Fill with the most frequent value or a specific value (e.g., False for "Has Table booking").
df['Has Table booking'].fillna(False, inplace=True)

# After handling missing values, re-check for any remaining missing values:
print(df.isnull().sum())

```

```python
df['Cuisines'] = df['Cuisines'].astype('category')
df['Currency'] = df['Currency'].astype('category')

df['Latitude'] = pd.to_numeric(df['Latitude'])
df['Longitude'] = pd.to_numeric(df['Longitude'])

# Check the distribution of the target variable
df['Aggregate rating'].value_counts()

```

```python
# Visualize the distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Aggregate rating', data=df)
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')
plt.show()


```

```python
# Calculate class proportions
class_proportions = df['Aggregate rating'].value_counts(normalize=True)
print(class_proportions)
```

```python
# 4. Analyze the distribution of the target variable
sns.histplot(df['Aggregate rating'], kde=True)
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')
plt.show()

```
